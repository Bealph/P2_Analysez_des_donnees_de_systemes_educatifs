## **Partie 4 : Filtrer et traiter les années les plus récentes dans le dataset**

**FIltrons les années recentes du dataframe**

```python
# appliquer la fonction filter 
# Selectionner les années allant de 2010 à 2016
df_filtre_annees_recentes = df_filtrer.filter(['Country Name', 'Country Code', 'Indicator Name','2010', '2011', '2012', '2013', '2014', '2015', '2016'])
df_filtre_annees_recentes
```

**Créons une colonne synthétique qu'on nomme 'Valeur moyenne' sur les années les plus récentes (2010-2016)**

```python
df_filtre_annees_recentes['Valeur_moy(synth)'] = df_filtre_annees_recentes[['2010', '2011', '2012', '2013', '2014', '2015', '2016']].mean(axis=1)
```

```python
df_filtre_annees_recentes.head()
```

**Créons un dataframe ayant en colonne le pays, le nom et le code des indicateurs, ainsi que notre colone synthétique**

```python
#Appliquons un filtre pour recuperer les colonnes qui nous interesse
df_col_filtrer = df_filtre_annees_recentes.filter(['Country Name', 'Country Code', 'Indicator Name', 'Valeur_moy(synth)'])
df_col_filtrer
```

**Créons notre dataframe post analyse**

```python
#Appliquons la fonction Pivot_table
data = df_col_filtrer.pivot_table(index=['Country Name', 'Country Code'], values = 'Valeur_moy(synth)', columns='Indicator Name', aggfunc='mean')
data.reset_index()
```
